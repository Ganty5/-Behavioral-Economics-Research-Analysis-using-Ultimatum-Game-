```python
#create a single binary variable for gender (0 for female, 1 for male)
Thesis['Gender'] = (Thesis['Gender'] == 'M').astype(int)
```

```python
Thesis.head(10)
```

```python
results = {}
```

```python
for rnd in range (1, 11):
    round_info = Thesis[Thesis['Round'] == 5]
    
```

```python
y = round_info['Response(value)']
```

```python
x = round_info[['Offers to Responders','Y_2021', 'Y_2020', 'Y_201916', 'Gender']]
```

```python
x = sm.add_constant(x)
```

```python
#perform the probit regression
```

```python
model = sm.Probit(y,x)
```

```python
result = model.fit()
```

```python
#store the result of the model
```

```python
results[rnd] = result.summary()
```
